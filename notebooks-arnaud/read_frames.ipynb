```python
import cv2
print(cv2.__version__)
```

```python
## cut frames at beginnig and end of videos

from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip

ffmpeg_extract_subclip("../data/videos/flying.MOV", 15, 145, targetname="../data/videos/flying_resized.MOV")
ffmpeg_extract_subclip("../data/videos/touching.MOV", 0, 120, targetname="../data/videos/touching_resized.MOV")
```

```python
## extract videos frames

for label in ['none', 'flying_resized', 'touching_resized']:
    print(label, "...")
    vidcap = cv2.VideoCapture(f"../data/videos/{label}.MOV")
    success,image = vidcap.read()
    count = 0
    success = True
    while success:
        cv2.imwrite(f"../data/images/{label}_frame{count}.jpg", image)     # save frame as JPEG file
        success,image = vidcap.read()
        print('Read a new frame: ', count, success)
        count += 1
```
